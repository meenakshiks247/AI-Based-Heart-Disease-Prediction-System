# Data Ingestion and Exploratory Analysis

This notebook demonstrates loading the heart disease dataset and performing initial exploratory data analysis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the heart disease dataset
df = pd.read_csv('../data/heart.csv')

# Display first 5 rows
df.head()
```

```python
# Check data info and null values
df.info()
```

```python
# Create a countplot of the target column
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='target', palette='magma')
plt.title('Heart Disease Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Target (0=No Disease, 1=Disease)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
