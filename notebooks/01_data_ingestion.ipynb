# Data Ingestion and Exploratory Analysis

This notebook demonstrates loading the heart disease dataset and performing initial exploratory data analysis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the heart disease dataset
df = pd.read_csv('../data/heart.csv')

# Print dataframe shape
print(f"Dataset shape: {df.shape}")
print()
```

```python
# Check data info and null values
df.info()
```

```python
# Create a countplot of the target column
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='target', hue='target', palette='magma', legend=False)
plt.title('Heart Disease Distribution: Healthy (0) vs. At Risk (1)', fontsize=14, fontweight='bold')
plt.xlabel('Target', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
